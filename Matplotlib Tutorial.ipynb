````python
import matplotlib.pyplot as plt
# pyplot is the graphing tool
plt.plot([1,2,3,4],[5,2,7,10])
# we are plotting 4 points (1,5),(2,2),(3,7),(4,10)
plt.show()
# keeps your graph on the screen
````

#### Creating a Legend

````python
x = np.arange(1,5)
plt.plot(x,x,x,x+1,x,x-1)
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(["Medium", "Fast", "Slow"]) # first label corresponds to the first pair of data
# legend function takes in an array of values, this array of value is a label for each line
plt.show()
````

````python
# Setting the legend inside the plot.
x = np.arange(1,5)
plt.plot(x,x, label = "Medium")
plt.plot(x,x+1, label= "Fast")
plt.plot(x,x-1, label = "Slow")
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend()
plt.show()
````

````python
# Setting the location(loc) of a legend upper, lower, lower right etc
x = np.arange(1,5)
plt.plot(x,x, label = "Medium")
plt.plot(x,x+1, label= "Fast")
plt.plot(x,x-1, label = "Slow")
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") #best takes into account your data and finds a location for you
plt.show()
````

#### Colour Customization

````python
# Every plot function has its own colour attribute
x = np.arange(1,5)
#color abbreviations: bcgkmrwy - blue, cyan green black,magenta, red,white,yellow
plt.plot(x,x, label = "Medium", color ="y")
plt.plot(x,x+1, label= "Fast", color ="g") # colour abbreviation g is green, you can also use rgb, hex value 
plt.plot(x,x-1, label = "Slow", color="r")
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") 
plt.show()
````

````python
# the third attribute can also be used to specify color
x = np.arange(1,5)
plt.plot(x,x, "y", label = "Medium")
plt.plot(x,x+1, label= "Fast", color ="g")
plt.plot(x,x-1, "#FF00FF", label = "Slow")
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") 
plt.show()
````

#### Line Customization
There are currently four line styles in Matplotlib:
* Dash "-" (solid line)
* Dash Dash "--" (dash line)
* Dash Dot "-." (dash dot line)
* Colon ":" (dotted line)

````python
# the third attribute can also be used to specify the line style
x = np.arange(1,5)
plt.plot(x,x, ":", label = "Medium")
plt.plot(x,x+1,label= "Fast", color ="g", linestyle = "-.")
plt.plot(x,x-1, label = "Slow", linestyle="--")
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") 
plt.show()
````

````python
# the third attribute is use for customization.
# we can specify the color and line style inside that attribute
x = np.arange(1,5)
plt.plot(x,x, "y--", label = "Medium") #yello dashed line
plt.plot(x,x+1,"-.",label= "Fast", color ="g")
plt.plot(x,x-1, "m:", label = "Slow") #magenta dotted line
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") 
plt.show()
````

#### Marker Customization

Marker style can also be used to customize your graphs. There are tons of marker abbreviations , but I will list some here :
* squares "s"
* pentagon "p"
* point "."
* circle "o"
* plus "+" 
* star "*"
* diamond "d"

````python
# the markers can also be specified in the third attribute
x = np.arange(1,5)
plt.plot(x,x, "p", label = "Medium") #pentagon marker
plt.plot(x,x+1,label= "Fast", color ="g", marker=".") #point marker
plt.plot(x,x-1,label = "Slow", marker="*")  #star marker
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") 
plt.show()
````

````python
# combining the marker attribute with line style and color
x = np.arange(1,5)
plt.plot(x,x, "y-p", label = "Medium") #yellow solid line with pentagon marker
plt.plot(x,x+1,"g-..",label= "Fast", color ="g") #green dash dot line with point marker
plt.plot(x,x-1, "r:+", label = "Slow")  #red dotted line with plus marker
plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")
plt.title("This is the title of the graph")
plt.legend(loc = "best") 
plt.show()
````

#### Multiple Figures in a Graph
So far, we had multiple charts in the same figure. Here we will try to create multiple figures for each chart and print them out in the same window.

We start by associating our entire window to a variable called fig.

fig =plt.figure()

To add multiple graph to your figure (fig), we create the following subplots. 

chart1 = fig.add_subplot()   
chart2 = fig.add_subplot()  
chart3 = fig.add_subplot()  

Inside the add_subplot function, we have to specify the position of the figure. It takes in 3 parameters, the first 2 being the window size and the last paramter is the position. The first two parameters contitues the row and the column. For example, if your window size is 2,1 (2X1), we have two position to place our graph. We can place the subplot in position 1 or 2.

|  | 
|--|
|1 |
|2 |

But we have 3 charts, so we will need a bigger window. Let's say we want our window size to be 3,1 (3X1), we will have 3 position to place our 3 charts. It will look something like this:


|  |
|--|
|1 |
|2 |
|3 |

Now, we can speicify each of the charts in each of the rows, like this:
chart1 = fig.add_subplot(3,1,1)  *chart 1 in position 1*  
chart2 = fig.add_subplot(3,1,2) *chart2 in position 2*  
chart3 = fig.add_subplot(3,1,3) *chart2 in position 3*

**Extra**  
Let's say our window size was 4,1 (4x1) , where would position 3 be?  
In this case, our figure will look something like this:  

|  |  |
|--|--|
|1 |2 |
|3 |4 |

Therefore, position 3 will be in the lower left hand corner.


````python
fig = plt.figure() 
chart1 = fig.add_subplot(3,1,1) 
chart2 = fig.add_subplot(3,1,2) 
chart3 = fig.add_subplot(3,1,3)

x=np.arange(1,5)
chart1.plot(x,x, "y-p", label = "Medium")
chart2.plot(x,x+1,"g-..",label= "Fast") 
chart3.plot(x,x-1, "r:+", label = "Slow")
plt.legend(loc="best")
plt.grid(True)
plt.axis([0,5,-1,13])
plt.show()

````

#### More Color Customization
Here we will learn to change background color of the figure, becakground color that is inside the axis, tick marks' color as well as the spines color. Spines represent the borderline of the chart and there are four in total : top, left, right and bottom.

````python
fig = plt.figure(facecolor="c") #background color of the figure is set to cyan
chart1 = fig.add_subplot(1,1,1, facecolor="y") #color inside the axis is set to yellow
x = np.arange(1,5)
chart1.plot(x,x,"g-p")
chart1.tick_params(axis="x",color="w") #tick marks on the x-axis of are set to white
chart1.tick_params(axis="y",color="w") #tick marks on the y-axis of are set to white
#chart1.spines["top"].set_color("w") #sets the top spine to white
#You can manually set all four of your spines or we we can use a for loop to cover all four spines
for spine in chart1.spines:
    chart1.spines[spine].set_color("w")
plt.show()
````
